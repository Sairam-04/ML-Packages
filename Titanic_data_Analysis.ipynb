For Cabin column we have 687 missing values and it is also not useful for prediction so we can drop Cabin column

```python
data['Cabin'].isnull().sum()
```

```python
data = data.drop(labels='Cabin',axis=1)
data.head()
```

Age column has 177 missing values and the type is int so we can replace all the NAN values with the mean.

```python
data['Age'].isnull().sum()
```

```python
mean_Age = data['Age'].mean()
print(mean_Age)
```

```python
from math import floor
data['Age'] = data['Age'].fillna(floor(mean_Age))
```

```python
data.isnull().sum()
```

```python
data.Age.head(10)
```

```python
data.isnull().sum()
```

For Embarked column we have 2 NAN values and the type of column is object so we cannot replace it with mean we can replace it by using mode. For Categorical data we should use mode to replace NAN values

```python
data['Embarked'].unique()
```

```python
data['Embarked'].value_counts()
```

We can see that in Embarked column we can replace NAN values with S because it has repeated more times.

```python
data['Embarked'] = data['Embarked'].fillna('S')
data['Embarked'].isnull().sum() 
```

```python
data.isnull().sum()
```

```python
data.Name.unique()
```

We don't need the Name column so we can drop the Name column

```python
data = data.drop('Name',axis=1)
```

```python
data.head()
```

We can also remove Ticket column

```python
data = data.drop('Ticket',axis=1)
data.head()
```
